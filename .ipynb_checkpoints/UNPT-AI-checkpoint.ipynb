```python
model = "gpt-4o"

userName = "이정호"
userAge = 21
userGender = "남성"
userMajor = "ERICA 공학대학 스마트융합공학부 스마트ICT융합전공"
userHobby = "코딩"
userCountry = "japan, korea, china"

userProfile = f"{userName}은(는) {userAge}살이고, {userGender}이며, {userMajor}을(를) 전공하고 있습니다. {userHobby}을(를) 좋아하며, {userCountry}에 있는 대학을 찾고 있습니다."
messages = [
    {
        "role": "system",
        "content": (
            "University Classification Prompt:\n\n"
            "Based on the provided university data, provide leisure, weather, location, and major information for the city where each university is located. Follow these criteria to classify each category:\n\n"
            "Leisure activities:\n\n"
            "- Classify as 'Extroverted (E)' if the city mainly features active facilities such as amusement parks, landmarks, cafes, bars, sports stadiums, etc.\n"
            "- Classify as 'Introverted (I)' if the city mainly features calm facilities such as libraries, cinemas, bookstores, etc.\n\n"
            "Weather:\n\n"
            "- Classify as 'Sunny (S)' if the city generally has warm weather with temperatures frequently above 15°C.\n"
            "- Classify as 'Freeze (F)' if the city generally has cold weather with temperatures frequently below 15°C.\n\n"
            "Location:\n\n"
            "- Classify as 'City (C)' if the city's population exceeds 100,000.\n"
            "- Classify as 'Town (T)' if the city's population is 100,000 or less.\n\n"
            "Major:\n\n"
            "- Classify as 'Liberal arts (L)' if the university is known for liberal arts, humanities, or business.\n"
            "- Classify as 'Nature (N)' if the university is known for science, technology, engineering, or mathematics.\n\n"
            "Ensure each university is classified with exactly one option for each category (E/I, S/F, C/T, L/N).\n\n"
            "Response format:\n\n"
            "대학: [University Name]\n"
            "UNPT: [E | I][S | F][C | T][L | N]\n"
            "여가정보: [Extroverted | Introverted]\n"
            "지리정보: [Sunny | Freeze]\n"
            "지역정보: [City | Town]\n"
            "전공정보: [Liberal arts | Nature]\n"
            "도시: [City Name]\n"
            "추천 시설: [Recommended Facility 1], [Recommended Facility 2], [Recommended Facility 3]\n\n"
            "Initial Question Generation:\n\n"
            "Based on the provided criteria, generate an initial open-ended question in Korean to classify the user's preferences. \n"
            "It doesn't give you any options to choose from.\n"
            "question is very creativity and diversity. Continue asking up to three open-ended questions based on the user's responses. After receiving the responses, classify the UNPT and recommend three universities that match the determined UNPT.\n"
            "Please return only questions\n\n"
            "End of prompt."
        )
    },
    {
        "role": "user",
        "content": f"이 유저의 프로필 데이터와 유사한 대학을 추천을 해주기 위한 정보를 얻기 위한 질문인데 3개만 할 수 있어. {userProfile}"
    }
]

response = openai.ChatCompletion.create(
    model= model,
    messages=messages,
    temperature=1,
    max_tokens=1000,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0
)

print(response.choices[0].message['content'])
```
